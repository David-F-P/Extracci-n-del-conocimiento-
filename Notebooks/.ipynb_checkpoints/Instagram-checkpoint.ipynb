```python
import pandas as pd
```

```python
import random
```

```python
# 1. Definimos la cantidad de filas 
cantidad = 1000000
```

```python
# 2. Conjunto de datos base reales
influencers_pool = ["@travel_veracruz", "@playa_MX", "@mexico_explorar", "@vida_estilo", "@paraiso_semanal", "@cancun_vacaciones"]
tipos_post_pool = ["foto", "video", "reel", "carrusel"]
fechas_pool = ["2026-05-01", "2026-05-10", "2026-05-15", "2026-05-20", "2026-05-24"]
```

```python
# 3. Creación del DataFrame de 1,000,000 de registros usando comprensión de listas
df_instagram = pd.DataFrame({
    "influencer": [random.choice(influencers_pool) for _ in range(cantidad)],
    "likes": [random.randint(100, 50000) for _ in range(cantidad)],  # Rango de likes realista
    "comentarios": [random.randint(5, 2500) for _ in range(cantidad)],  # Rango de comentarios realista
    "fecha": [random.choice(fechas_pool) for _ in range(cantidad)],
    "tipo_post": [random.choice(tipos_post_pool) for _ in range(cantidad)]
})
```

```python
# Definimos la cantidad (1 millón de registros)
cantidad = 1000000
```
